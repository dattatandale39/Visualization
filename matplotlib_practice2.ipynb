```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('gas price.csv')
df
```

```python
df.shape
```

```python
df=df.head(100)
df.info()
```

```python
plt.figure(figsize=(9,5),dpi=120)

plt.plot(df.Year,df.Canada,'r.--',label='Canada')
plt.plot(df.Year,df.Australia,'g-->',label='Australia')
plt.plot(df.Year,df.USA,'b^--',label='USA')
plt.title('gas price over time(USD)')
plt.xticks(df.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,5),dpi=120)

for country in df:
    if country !='Year':
        plt.plot(df.Year,df[country],'.-',label=country)

plt.title('gas price over time(USD)')
plt.xticks(df.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,5),dpi=120)

selected_countries=['USA','Canada','Mexico']
for country in df:
    if country in selected_countries:
        plt.plot(df.Year,df[country],'.-',label=country)

plt.title('gas price over time(USD)')
plt.xticks(df.Year[::3])
plt.xlabel('Year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,5),dpi=120)

plt.plot(df.Year,df.Canada,'r.--',label='Canada')
plt.plot(df.Year,df.Australia,'g-->',label='Australia')
plt.plot(df.Year,df.USA,'b^--',label='USA')
plt.title('gas price over time(USD)',fontdict={'fontweight':'bold','fontsize':22})
plt.xticks(df.Year[::3].tolist()+[2011])
plt.xlabel('Year',fontdict={'fontweight':'bold','fontsize':12})
plt.ylabel('USD',fontdict={'fontweight':'bold','fontsize':12})
plt.legend()
plt.savefig('new gas.png',dpi=130)
plt.show()
```
